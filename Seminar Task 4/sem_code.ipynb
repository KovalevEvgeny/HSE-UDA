```python
from nltk import DependencyGraph
import codecs

processed_sentences = []
sentence = []
for line in codecs.open('data.conll', 'r', 'utf-8'):
    if len(line) == 1:
        processed_sentences.append(sentence)
        sentence = []
    else:
        word = line.split("\t")
        sentence.append(word)
```

```python
deps = []
for sentence in processed_sentences:
    s = u''
    for line in sentence:
        s += u"\t".join(line) + u'\n'
    deps.append(s)

for sent_dep in deps:
    graph = DependencyGraph(tree_str=sent_dep)
    for triple in graph.triples():
        for elem in triple:
            print (elem[0] if isinstance(elem, tuple) else elem,)
        print()
    print()
    tree = graph.tree()
    print(tree.pretty_print())
```

```python
for sent_dep in deps:
    verbs = {}
    for t in sent_dep.split('\n'):
        if len(t) > 1:
            splt = t.split('\t')
            if splt[3] == 'VERB':
                verbs[splt[0]] = [splt[1]]

    for t in sent_dep.split('\n'):
        if len(t) > 1:
            splt = t.split('\t')
            if splt[7] in ['dobj', 'nsubj']:
                if splt[6] in verbs:
                    verbs[splt[6]].append(splt[1])
    for t in verbs.items():
        for elem in t:
            print (elem,)
        print()
```
